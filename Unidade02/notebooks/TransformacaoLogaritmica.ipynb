```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Gera um dataset skewed 
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000) 

# Cria o dataframe
df = pd.DataFrame({'Original': data, 'Log': np.log1p(data)})

# Plota histogramas
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Original'], kde=True, bins=30)
plt.title('Original Data Histograma')

plt.subplot(1, 2, 2)
sns.histplot(df['Log'], kde=True, bins=30)
plt.title('Log Data Histograma')

plt.tight_layout()
plt.show()

# Plota boxplots
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Original'])
plt.title('Original Data Boxplot')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Log'])
plt.title('Log Data Boxplot')

plt.tight_layout()
plt.show()
```
